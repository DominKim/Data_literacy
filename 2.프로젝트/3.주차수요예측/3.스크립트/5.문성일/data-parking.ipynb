```python
train=pd.read_csv('/kaggle/input/parking-data/train.csv')
test=pd.read_csv('/kaggle/input/parking-data/test.csv')
sub = pd.read_csv('/kaggle/input/parking-data/sample_submission.csv')
```

```python
all_data = pd.concat([train,test])
all_data.head()
```

```python
all_data['자격유형'].unique()
```

```python
#all_data['단지코드'].unique() # index 역할.
```

```python
all_data2 = all_data.drop(columns=['단지코드','등록차량수'],axis=1)
all_data2
```

```python
all_data2[['임대보증금','임대료']].head(20)
```

```python
c=all_data2.columns[all_data2.dtypes=='object']
c
# 임대보증금, 임대료는 문자형으로 인코딩하면 안된다. -> 숫자로 변환.
# 특수문자가 들어가 있음.
```

```python
# 특수문자를 np.nan으로 변경하자.
all_data2[['임대보증금','임대료']]=all_data2[['임대보증금','임대료']].replace('-',np.nan)
all_data2
```

```python
all_data2[['임대보증금','임대료']]=all_data2[['임대보증금','임대료']].astype(float)

# int로 하면 오류가 난다. why? -> 결측치(missing_value)가 존재하기 때문.
```

```python
c=all_data2.columns[all_data2.dtypes=='object']
c
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for i in c:
    all_data2[i] = le.fit_transform(list(all_data2[i]))
    
all_data2
```

```python
train2 = all_data2[:len(train)]
test2 = all_data2[len(train):]
```

```python
from catboost import CatBoostRegressor
cbr = CatBoostRegressor(learning_rate=0.01,iterations=10000)
cbr.fit(train2,train['등록차량수'])
result = cbr.predict(test2)

# 과대적합 하지 않도록 CV
# Test 데이터가 너무 적어서 과대적합 막아야 한다.
```

```python
# 2가지가 필요.
# test의 단지코드를 기준으로 묶어야 한다.
# result 값이 필요.|

df = pd.DataFrame({'code':test['단지코드'],'result':result})
df
```

```python
# 단지코드별 등록차량수가 같음. -> 평균으로 해야 함.
# 단지코드별로 등록차량수가 다양했다면 -> 합으로 해야 함.
# 단지코드별로 등록차량수가 같기 때문에 평균, median으로 설정함.
```

```python
df = df.groupby(['code'])['result'].mean()
df
```

```python
# sub와 비교했을 때,
# 순서가 다르다. -> sub에 맞게 index를 설정.
```

```python
sub=pd.merge(sub,df,on='code',how='left')
sub
```

```python
sub = sub[['code','result']]
sub.columns = ['code','num']
sub
```
